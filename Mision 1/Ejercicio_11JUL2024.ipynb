```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import datetime as dt
import calendar

calendar

Fecha_acctual = dt.datetime.now()
print("La fecha hoy es: ",Fecha_acctual, "\n")

# Create a dataset
value_personas = ['Pedro Pérez', 'Juan Sánchez', 'Guillermo Buitrago', 'Ana María', 'Sara Sofia', 'Juliana Garcia', 'Efraín Ortega', 'Leo Torres', 'Patricia Rodríguez', 'Tom y Jerry']
value_edad = [40, 60, 70, 25, 45, 60, 30, 53, 62,65] #Edad en años
value_peso = [70, 78, 91.5, 58, 65, 56, 75, 82, 63, 90] # [peso en kg]
value_altura = [1.76, 1.87, 1.69, 1.62, 1.65, 1.75, 1.65, 1.70, 1.65, 1.90] # [Esatura en m]

data_dict = {"Nombre":value_personas,
             "Edad":value_edad,
             "Peso":value_peso,
             "Estatura":value_altura} #,
              #"BMI":value_bmi_list}
df = pd.DataFrame(data_dict)
df["IMC"] = round(df["Peso"] / df["Estatura"]**2,2)
df
#df.describe()
#print(df)

```

```python
# Generate synthetic data
#np.random.seed(42)  # for reproducibility

data = {"Edad":value_edad,
        "Peso":value_peso,
        "Estatura":value_altura,
        "IMC":df["IMC"]}

df = pd.DataFrame(data)
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Print the correlation matrix
print("Correlation Matrix prueba: \n")
print(correlation_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix prueba')
plt.show()
```

```python
def condicionsalud (IMC):
  if IMC < 18.5:
    return "Bajo peso"
  elif 18.5 <= IMC < 25:
      return "Peso normal"
  elif 25 <= IMC < 30:
     return "Sobrepeso"
  else:
     return "Obesidad"

df["Estado_Salud"] = df["IMC"].apply(condicionsalud)
print(df)

#df.describe()
df_stats = df.describe()
df_stats
```
